```python
%matplotlib inline
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
dfindex = pandas.read_csv('data/index_html.csv')
dfindex.keys()

dfchicle = pandas.read_csv('data/chicle_html.csv')
dfchicle.keys()

dfcaramelo = pandas.read_csv('data/caramelo_html.csv')
dfcaramelo.keys()

dfmarshmallow = pandas.read_csv('data/marshmallow_html.csv')
dfmarshmallow.keys()

dflollipop = pandas.read_csv('data/lollipop_html.csv')
dflollipop.keys()

dfsugus = pandas.read_csv('data/sugus_html.csv')
dfsugus.keys()
```

### Carga de todos los valores en un mismo gr√°fico:

```python
plt.figure(figsize=(30, 15))

_ = plt.hist(dfindex['seconds'], histtype='step', fill=False)
_ = plt.hist(dfchicle['seconds'], histtype='step', fill=False)
_ = plt.hist(dfcaramelo['seconds'], histtype='step', fill=False)
_ = plt.hist(dflollipop['seconds'], histtype='step', fill=False)
_ = plt.hist(dfmarshmallow['seconds'], histtype='step', fill=False)
_ = plt.hist(dfsugus['seconds'], histtype='step', fill=False)

plt.xlabel('seconds')
plt.ylabel('visits')
```

### Datos de caramelo duro

```python
_ = plt.hist(dlollipop['seconds'])
_ = plt.hist(dcaramelo['seconds'])

plt.suptitle('Caramelo duro')
plt.ylim(0,300)
plt.xlim(0,45)
```

### Caramelo duro VS Caramelo blando agrupado y con formateo de gr√°ficas

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 10))

fig.subplots_adjust(left=0.1, right=0.95, bottom=0.15)

ax[0].set_title('Caramelo duro')
ax[0].set_xlabel('seconds')
ax[0].set_ylabel('visits')
ax[0].set_ylim(0,300)
ax[0].set_xlim(0,45)
ax[0].hist(dlollipop['seconds'], histtype='step', fill=False, label='Lollipop')
ax[0].hist(dcaramelo['seconds'], histtype='step', fill=False, label='Caramelo')
ax[0].legend(prop={'size': 15})

ax[1].set_title('Caramelo blando')
ax[1].set_xlabel('seconds')
ax[1].set_ylabel('visits')
ax[1].set_ylim(0,300)
ax[1].set_xlim(0,45)
ax[1].hist(dmarshmallow['seconds'], histtype='step', fill=False, label='Marhsmallow')
ax[1].hist(dchicle['seconds'], histtype='step', fill=False, label='Chicle')
ax[1].hist(dsugus['seconds'], histtype='step', fill=False, label='Sugus')
ax[1].legend(prop={'size': 15}) 

```
